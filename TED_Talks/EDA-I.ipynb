```python
def basic_clean(text):
    text = text.lower()
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r"[^a-z0-9?!'.]", ' ', text)
    text = re.sub(r'\s+', ' ', text)
    text = text.strip()

    return text
```

```python
df = df.copy()
```

```python
df['clean_description'] = df['description'].apply(basic_clean)
df['clean_transcript'] = df['transcript'].apply(basic_clean)
```

```python
df
```

```python
df['desc_word_count'] = df['clean_description'].str.split().str.len()
df['script_word_count'] = df['clean_transcript'].str.split().str.len()
```
